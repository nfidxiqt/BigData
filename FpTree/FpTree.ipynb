```python
class TreeNode:
  def __init__(self,name,num,parent):
    self.name = name
    self.num = num
    self.parent = parent
    self.children = {}

  def add(self, num):
    self.num += num
```

```python
###2###
def Check_First_Set(datalist,sup):
  numDict = {}
  supportSet = set()
  for line in datalist:
    for item in line:
      numDict[item] = numDict.get(item, 0) + 1
      if (numDict[item] >= sup):
        supportSet.add(frozenset([item]))

  return list(supportSet)
```

```python
###0###
def Get_Second_Set(Set):
  length = len(Set)
  newSet = set()
  for item1 in range(length):
    for item2 in range(item1 + 1,length):
      newSet.add(Set[item1] | Set[item2])

  return list(newSet)
```

```python
###8###
def Get_New_Set(Set,k):
  length = len(Set)
  newSet = set()
  for item1 in range(length):
    for item2 in range(item1 + 1, length):
      if(sorted(list(Set[item1]))[:k - 2] == sorted(list(Set[item2]))[:k - 2]):
        unionSet = Set[item1] | Set[item2]
        subset = list(combinations(unionSet,k - 1))
        for i in subset:
          if (set(i) not in Set):
            break
        else:
          newSet.add(unionSet)

  return list(newSet)
```

```python
def Creat_Data_List(nodeList):
  newData = {}
  for node in nodeList:
    line = []
    num = node.num
    node = node.parent
    while node.name != 'Null':
      line.append(node.name)
      node = node.parent
    if (len(line) > 0):
      newData[frozenset(line)] = num

  return newData
    
```

```python
def Creat_Tree(newData,headDict,numDict):
  retNode = TreeNode('Null', 1, None)
  for line,num in newData.items():
    line = list(line)
    line.sort(key=lambda item: (numDict[item], item),reverse=True)
    Insert_Node(line,retNode,headDict,num)

  return retNode,headDict
```

```python
def Insert_Node(line,parentNode,headDict,num):
  item = line[0]
  if (item in parentNode.children):
    parentNode.children[item].add(num)
  else:
    newNode = TreeNode(item,num,parentNode)
    parentNode.children[item] = newNode
    headDict[item].append(newNode)
  if (len(line) > 1):
    Insert_Node(line[1:],parentNode.children[item],headDict,num)

  return
    
```
